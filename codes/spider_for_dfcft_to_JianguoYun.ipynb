```python
from bs4 import BeautifulSoup
import requests
import urllib
import pandas as pd
```

```python
html_doc = urllib.request.urlopen('http://data.eastmoney.com/report/hyyb.html#dHA9MCZjZz0wJmR0PTImaHk9NDY1JnBhZ2U9MQ==').read()
```

```python
#解析html文件
soup = BeautifulSoup(html_doc, "lxml")
```

```python
#找出所有script标签,并文本化
soup2 = str(soup.find_all('script'))
```

```python
#分离出有效数据，将其转化为list，方法不稳定！
soup3 = soup2.split(':["')[1].split('"]')[0].split('","')
```

```python
#将数据转化为DataFrame可用格式，处理文本
soup4 = []
for i in range(0, len(soup3)-1):
    soup4.append(soup3[i].replace('&sbquo;', '，').replace('&quot;', '\"').split(','))
```

```python
#将数据转化为DataFrame
dataf = pd.DataFrame(soup4,)
```
