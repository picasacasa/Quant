```python
t1 = soup0.find_all('table')[0]
```

```python
# t1.find_all('tr')
```

```python
l1 = []
l2 = []
for i in range(0, len(t1.find_all('th'))):
    l2.append(t1.find_all('th')[i].text)
l1.append(l2)
l2 = []
for i in range(1, len(t1.find_all('tr'))):
    l2 = []
    for j in t1.find_all('tr')[i].find_all('td'):
        l2.append(j.text)
    l1.append(l2)

    
```
